### Random Choice MNIST Data Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets import fetch_openml
```

```python
mnist = fetch_openml('mnist_784')
mnist.data.shape, mnist.target.shape
```

```python
y_value = mnist.target.astype('float')

abd = []
for i in range(10):
    print('y_value =', i)
    print('y_value_length =',len(y_value[y_value==i]))
    abd.append(len(y_value[y_value==i]))
    
print()
print('total_y_value_length =', sum(abd))
print(min(abd))
```

```python
zero_idx = []
one_idx = []
two_idx = []
three_idx = []
four_idx = []
five_idx = []
six_idx = []
seven_idx = []
eight_idx = []
nine_idx = []

for idx, value in enumerate(y_value):
    if value == 0:
        zero_idx.append(idx)
    elif value == 1:
        one_idx.append(idx)
    elif value == 2:
        two_idx.append(idx)
    elif value == 3:
        three_idx.append(idx)
    elif value == 4:
        four_idx.append(idx)
    elif value == 5:
        five_idx.append(idx)
    elif value == 6:
        six_idx.append(idx)
    elif value == 7:
        seven_idx.append(idx)
    elif value == 8:
        eight_idx.append(idx)
    elif value == 9:
        nine_idx.append(idx)
```

```python
x_zero_value = mnist.data[zero_idx]
x_one_value = mnist.data[one_idx]
x_two_value = mnist.data[two_idx]
x_three_value = mnist.data[three_idx]
x_four_value = mnist.data[four_idx]
x_five_value = mnist.data[five_idx]
x_six_value = mnist.data[six_idx]
x_seven_value = mnist.data[seven_idx]
x_eight_value = mnist.data[eight_idx]
x_nine_value = mnist.data[nine_idx]
```

```python
for i in np.random.choice(6313, 10, replace=False):
    fig = plt.figure(figsize=(10, 10))
    
    plt.subplot(10,10,1)
    plt.imshow(x_zero_value[i].reshape(28, 28))
    plt.xticks([])
    plt.yticks([])
    
    plt.subplot(10,10,2)
    plt.imshow(x_one_value[i].reshape(28, 28))
    plt.xticks([])
    plt.yticks([])
    
    plt.subplot(10,10,3)
    plt.imshow(x_two_value[i].reshape(28, 28))
    plt.xticks([])
    plt.yticks([])
    
    plt.subplot(10,10,4)
    plt.imshow(x_three_value[i].reshape(28, 28))
    plt.xticks([])
    plt.yticks([])
    
    plt.subplot(10,10,5)
    plt.imshow(x_four_value[i].reshape(28, 28))
    plt.xticks([])
    plt.yticks([])
    
    plt.subplot(10,10,6)
    plt.imshow(x_five_value[i].reshape(28, 28))
    plt.xticks([])
    plt.yticks([])
    
    plt.subplot(10,10,7)
    plt.imshow(x_six_value[i].reshape(28, 28))
    plt.xticks([])
    plt.yticks([])
    
    plt.subplot(10,10,8)
    plt.imshow(x_seven_value[i].reshape(28, 28))
    plt.xticks([])
    plt.yticks([])
    
    plt.subplot(10,10,9)
    plt.imshow(x_eight_value[i].reshape(28, 28))
    plt.xticks([])
    plt.yticks([])
    
    plt.subplot(10,10,10)
    plt.imshow(x_nine_value[i].reshape(28, 28))
    plt.xticks([])
    plt.yticks([])
    
    plt.tight_layout()
    plt.show()
```

```python
ab = [x_zero_value,x_one_value,x_two_value,x_three_value,x_four_value
     ,x_five_value,x_six_value,x_seven_value,x_eight_value,x_nine_value]

for i in np.random.choice(6313, 10, replace=False):
    fig = plt.figure(figsize=(10, 10))
    
    for idx, k in enumerate(ab):

        plt.subplot(1,10,idx+1)
        plt.imshow(k[i].reshape(28, 28))
        plt.xticks([])
        plt.yticks([])

    plt.tight_layout()
    plt.show()
```
